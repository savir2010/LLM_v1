```python
loader = CSVLoader(file_path='/content/drive/MyDrive/Data/savirbasics_faqs.csv', source_column="prompt", encoding=encoding)
data_load = loader.load()
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings()
query_result = embeddings.embed_query('What is your refund policy')
query_result
```

```python
from langchain.vectorstores import FAISS
```

```python
vectordb = FAISS.from_documents(documents=data_load,embedding=embeddings)
```

```python
retriever = vectordb.as_retriever()
```

```python
retriever.get_relevant_documents("For how long is this course valid")
```

```python
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
```

```python
prompt_template = """Given the following context and a question, generate an answer based on this context only.
In the answer try to provide as much text as possible from "response" section in the source document context without making much changes.
If the answer is not found in the context, kindly state "I don't know." Don't try to make up an answer.

CONTEXT: {context}

QUESTION: {question}"""

PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
    )
```

```python
chain = RetrievalQA.from_chain_type(llm=llm,
                                    chain_type="stuff",
                                    retriever=retriever,
                                    input_key="query",
                                    return_source_documents=True,
                                    chain_type_kwargs={"prompt": PROMPT})
```

```python
chain("do you have javascipt coursessss")
```
